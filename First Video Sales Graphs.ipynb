```python
import pandas as pd
df = pd.read_csv('/users/bentonmaples/downloads/Video_Games_Sales_as_at_22_Dec_2016.csv')
print(df)
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(df['Global_Sales'])
plt.show()
```

```python
# The above graph should be a graph of total sales volume for all titles in the data set.  It is immeditately clear
# that there is a small percentage that are responsible for nearly all units sold.

plt.plot(df['Global_Sales'])
plt.ylabel('Units')
plt.title('Global Sales Volume')
plt.show()
```

```python
# The data visualization is improved with labels

plt.plot(df['Global_Sales'])
plt.ylabel('Units')
plt.title('Global Sales Volume')
plt.xlim([-1, 170])
plt.show()
```

```python
# Should show sales violume of approximately top 1%
# Now will graph top .1%

plt.plot(df['Global_Sales'])
plt.ylabel('Units')
plt.title('Global Sales Volume')
plt.xlim([-1, 17])
plt.show()
```

```python
# Now will look at top 1000 titles

plt.plot(df['Global_Sales'])
plt.ylabel('Units')
plt.title('Global Sales Volume')
plt.xlim([0, 1000])
plt.show()
```

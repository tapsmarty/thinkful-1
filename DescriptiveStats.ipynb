```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
distribution1 = np.random.binomial(20,.5, 100)
d1_mean = np.mean(distribution1)
d1_std = np.std(distribution1)

plt.hist(distribution1)
plt.plot([d1_mean, d1_mean], [0,30], label = "mean", color="red")
plt.plot([d1_std, d1_std], [0,30], label = "std dev", color="green")

plt.legend()
plt.show()
```

```
distribution2 = np.random.exponential(4, 100)
d2_mean = np.mean(distribution2)
d2_std = np.std(distribution2)

plt.hist(distribution2)
plt.plot([d2_mean, d2_mean], [0,30], label = "mean", color="red")
plt.plot([d2_std, d2_std], [0,30], label = "std dev", color="green")

plt.legend()
plt.show()
```

```
distribution4 = np.random.standard_normal(100)
d4_mean = np.mean(distribution4)
d4_std = np.std(distribution4)

plt.hist(distribution4)
plt.plot([d4_mean, d4_mean], [0,30], label = "mean", color="red")
plt.plot([d4_std, d4_std], [0,30], label = "std dev", color="green")

plt.legend()
plt.show()
```

```
distribution5 = np.random.beta(1, 4, 100)
d5_mean = np.mean(distribution5)
d5_std = np.std(distribution5)

plt.hist(distribution5)
plt.plot([d5_mean, d5_mean], [0,30], label = "mean", color="red")
plt.plot([d5_std, d5_std], [0,30], label = "std dev", color="green")

plt.legend()
plt.show()
```

```
distribution6 = np.random.vonmises(1,4, 100)
d6_mean = np.mean(distribution6)
d6_std = np.std(distribution6)

plt.hist(distribution6)
plt.plot([d6_mean, d6_mean], [0,30], label = "mean", color="red")
plt.plot([d6_std, d6_std], [0,30], label = "std dev", color="green")

plt.legend()
plt.show()
```

```
# With exception of expononential and vonmises, all distributions seem to be normal with a mean falling in the 
# middle of a normally distributed graph and the standard deviation significcantly lower than the mean.  In the
# exponential and vonmises distributiuons, values cluster on the left side of the graph and the standard deviation
# is large relative to the mean.
```

```
var1 = np.random.normal(5, .5, 100)
plt.hist(var1)
plt.show
```

```
var2 = np.random.normal(10, 1, 100)
plt.hist(var2)
plt.show
```

```
newvar = var1 + var2
new_mean = np.mean(newvar)
new_std = np.std(newvar)

plt.hist(newvar)
plt.plot([new_mean, new_mean], [0,30], label = "mean", color="red")
plt.plot([new_std, new_std], [0,30], label = "std dev", color="green")

plt.legend()
plt.show()
```

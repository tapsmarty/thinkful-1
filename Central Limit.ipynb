```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000) 

# Letâ€™s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

sample1_mean = np.mean(sample1)
sample2_mean = np.mean(sample2)

plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print("Sample 1 mean (100): " + str(sample1_mean))
print("Sample 2 mean (100): " + str(sample2_mean))
```

```python
# Increasing the limit to 1000 should give more accurate representation, but not much will change
```

```python
sample3 = np.random.choice(pop1, 1000, replace=True)
sample4 = np.random.choice(pop2, 1000, replace=True)

sample3_mean = np.mean(sample3)
sample4_mean = np.mean(sample4)

plt.hist(sample3, alpha=0.5, label='sample 3') 
plt.hist(sample4, alpha=0.5, label='sample 4') 
plt.legend(loc='upper right') 
plt.show()
print("Sample 3 mean (1000): " + str(sample3_mean))
print("Sample 4 mean (1000): " + str(sample4_mean))
```

```python
sample5 = np.random.choice(pop1, 20, replace=True)
sample6 = np.random.choice(pop2, 20, replace=True)

sample5_mean = np.mean(sample5)
sample6_mean = np.mean(sample6)

plt.hist(sample5, alpha=0.5, label='sample 5') 
plt.hist(sample6, alpha=0.5, label='sample 6') 
plt.legend(loc='upper right') 
plt.show()
print("Sample 5 mean (20): " + str(sample5_mean))
print("Sample 6 mean (20): " + str(sample6_mean))
```

```python
pop1 = np.random.binomial(10, 0.3, 10000)
sample7 = np.random.choice(pop1, 100, replace=True)
sample8 = np.random.choice(pop1, 100, replace=True)

plt.hist(sample7, alpha=0.5, label='sample 7') 
plt.hist(sample8, alpha=0.5, label='sample 8') 
plt.legend(loc='upper right') 
plt.show()

print(sample7.mean())
print(sample8.mean())
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(sample7, sample8, equal_var=False))
```

```python
pop1 = np.random.binomial(10, 0.4, 10000)
sample9 = np.random.choice(pop1, 100, replace=True)
sample10 = np.random.choice(pop1, 100, replace=True)

plt.hist(sample9, alpha=0.5, label='sample 9') 
plt.hist(sample10, alpha=0.5, label='sample 10') 
plt.legend(loc='upper right') 
plt.show()

print(sample9.mean())
print(sample10.mean())
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(sample9, sample10, equal_var=False))
```

```python
# T-statistic increases and p-value decreases as p changes from .3 to .4.  
# T-statistic increases because there appears less noise in data with more predictable likelyhood of
# k success given n tries.  As p increases from .3 to .4, data appears less noisy.
#
# p-value decreases as probability of n success in k tries increases because the samples are more likely
# to be similar as p increases in binomial
```
